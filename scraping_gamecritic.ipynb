```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
urls = pd.read_csv('Data2017.csv',header=0)
urls['Critic Score'].astype(float)
urls
```

```python
urlselect = urls[urls['Critic Score'] == 'tbd']
urlselect
```

```python
import re
from statistics import mean
testurl = urls.iloc[3]
plat = testurl['Game Platform'].lower().replace(' ','-')
title = re.sub(r'[\W_]', '-', testurl['Game Title'].lower())
titleresult = title.replace('--','-')
print(plat)
print(titleresult)

header= {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:62.0) Gecko/20100101 Firefox/62.0"}
urlcritic = 'https://www.metacritic.com/game/'+str(plat)+'/'+str(titleresult)+'/critic-reviews'
page=requests.get(urlcritic,headers=header)
soup = BeautifulSoup(page.text, 'html.parser')

container_grade = soup.find_all('div', class_ = 'review_grade')
container_count = int(soup.find_all('strong')[0].get_text().strip().replace(' Critic Reviews',''))
print(container_count)
game_grade = []
for i in container_grade:
    game_grade.append(i.get_text().strip())
game_grade = (game_grade[0:container_count])
game_grade = [int(i) for i in game_grade]
print(mean(game_grade))
```
